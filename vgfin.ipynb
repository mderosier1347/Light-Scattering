```python
import numpy as np
import cmath as cm
from collections import Counter
import json

w = float(input("Enter the wavelength: "))
NSID = int(input("Enter the number of dipoles along the major axes: "))
NLSID = int(input("Enter the number of fine structure devisions per dipole cell side: "))
RAD = float(input("Enter the particle symmetry semi-axes (z-axes radius): "))
alpha = float(input("Enter the incident theta (in degrees): "))
beta = float(input("Enter the value for phi (in degrees): "))
gamma = float(input("Enter the value for gamma (in degrees): "))
psi = float(input("Enter the polarization angle (in degrees): "))
MR = float(input("Enter the real part of the the index of refraction: "))
MIstring = (input("Enter the imaginary part of the index of refraction: "))
MI = float(MIstring)
ARstring = (input("Enter the aspect ratio (major axis/minor axis): ")) 
AR = float(ARstring)
COMMENTS = input("Enter the file comments: ")
OUTFILE = input("Enter the output filename: ")

def SD(W,MI):
    if MI == 0:
        return SD == 0
    else:
        return SD == (W/(2*np.pi*MI))
    return SD

Max = max(RAD, (RAD/AR))

TD = ((2*Max)/NSID)

if MI < 0:
    MI = -MI
M = complex(MR,MI)

DRF = 0.5*(NSID - 1)

LD = TD/NLSID

FRAC = 1/NLSID

NUSE = 0

INCNT = 0 

temp = 0

R = np.zeros((NUSE,3))
D = np.zeros((NUSE))

for i in range (0, NSID-1):
    for h in range (0, NSID-1):
        for g in range (0, NSID-1):
            x = (i - DRF)*TD
            y = (h - DRF)*TD
            z = (g - DRF)*TD
            for IX in range (0, NLSID - 1):
                for IY in range (0, NLSID - 1):
                    for IZ in range (0, NLSID - 1):
                        RF1 = x + LD*(0.5*(1.0 - NLSID) + IX)      
                        RF2 = y + LD*(0.5*(1.0 - NLSID) + IY)
                        RF3 = z + LD*(0.5*(1.0 - NLSID) + IZ)
                
                        INSIDE = np.sqrt((RF1/(RAD**2/AR**2)) + np.sqrt((RF2/(RAD**2/AR**2)) + np.sqrt(RF2/RAD**2)))
                        if 1 < INSIDE:
                            INCNT = INCNT + 1
                        
                        
                        
            if INCNT > 0:
                NUSE = NUSE + 1
                if NUSE/1000 < 0:
                    print (NUSE)
                    R[NUSE,1] = x #RF1 ? 
                R[NUSE,2] = y
                R[NUSE, 3] = z
                
                
                D[NUSE] = TD*(INCNT/(NLSID**3))**(1/3)
            INCNT = 0
print(type(R))
print(R)

#for i in range (0, NUSE):
    #R = np.array(R[i,1], R[i,2], R[i, 3])
    #D = np.array(D[i])
    #temp = temp + D(i)**3
            
print (COMMENTS)
print (NUSE)
print (w, alpha, beta, gamma, psi, RAD)
EPS = M*M
ER = EPS
EI = complex (EPS)
print (ER, EI, TD)


data=[]
data.append({'Comments': COMMENTS, 'Counter': NUSE, 'Wavelength': w, 'Incident theta angle': alpha, 'Value for phi': beta})
data.append({'Value for gamma': gamma, 'Polarization angle': psi, 'Particle symmetry semi-axes': RAD})
data.append({'EPS': str(EPS), 'ER': str(ER), 'EI': str(EI), 'TD': str(TD)})
data.append({'R-vector': np.ndarray.tolist(R)})
data.append({'D-vector': np.ndarray.tolist(D)})
#temp = temp + D(i)**3
data.append({'Number of dipoles along major axes': NSID, 'Number of fine structure devisions per dipole cell side': NLSID})
data.append({'Real part of index of refraction': MR, 'Imaginary part of index of refraction': MI})


with open('/Users/madisonderosier/Desktop/Research/' + OUTFILE + '.txt', 'w') as fp:
    variable = json.dumps(data, indent=4)
    fp.write(variable)
    

#file = open("/Users/madisonderosier/Desktop/Research/" + OUTFILE + ".txt","w")

#file.write(COMMENTS)
#file.write(str(NUSE))
#file.write(' {}{}\n'.format(" Wavelength ", w))
#file.write(str(w))
#file.write(' {}{}\n'.format(" Incident theta angle ", alpha))
#file.write(' {}{}\n'.format(" Value for phi ", beta))
#file.write(' {}{}\n'.format(" Value for gamma ", gamma))
#file.write(' {}{}\n'.format(" Polarization angle ", psi))
#file.write(' {}{}\n'.format(" Particle symmetry semi-axes ", RAD))
#file.write(str(EPS))
#file.write(str(ER))
#file.write(str(EI))
#file.write(str(TD))
#for i in range (0, NUSE):
    #file.write(' {}{}{}{}\n'.format(R[i,1], R[i,2], R[i, 3], D[i]))
    #temp = temp + D(i)**3
#file.write(' {}{}\n'.format(" Number of dipoles along major axes ", NSID))
#file.write(' {}{}\n'.format(" Number of fine structure devisions per dipole cell side ", NLSID))
#file.write(' {}{}\n'.format(" Real part of index of refraction ", MR))
#file.write(' {}{}\n'.format(" Imaginary part of the index of refraction ", MI))


#file.close()
    



```
